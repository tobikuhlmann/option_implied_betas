##### Get  constituent free float shares data for every constituent that has been in the S&P 500 historically

```python
n=test_ljc.shape[0]
for i in range(0, n):
     # check if file already exists, if true skip
    exists = os.path.isfile("C:/Users/uqdkx/Desktop/Free_Float_Shares/"+test_ljc.iloc[i][2]+".csv")
    if exists:
        print(i, ': ', test_ljc.iloc[i][2], ' already exists')
        continue
    else:
        print(i, ': ', test_ljc.iloc[i][2])
        file_exist = False
        try:
            timeseries, err = ek.get_data([test_ljc.iloc[i][2]], 
                                          ['TR.FreeFloat.Date', 'TR.FreeFloat'], 
                                          {'SDate':'1990-01-01', 'Frq':'D','EDate':'1998-12-31'})
            if timeseries.shape[0] > 0:
                timeseries.to_csv("C:/Users/uqdkx/Desktop/Free_Float_Shares/"+test_ljc.iloc[i][2]+".csv", index = True)
                file_exist = True
        except Exception as exc:
            if 'No data available for the requested date range' in str(exc):
                pass
            else:
                print(exc)

        time.sleep(1)
        try:
            timeseries, err = ek.get_data([test_ljc.iloc[i][2]], 
                                          ['TR.FreeFloat.Date', 'TR.FreeFloat'], 
                                          {'SDate':'1999-01-01', 'Frq':'D','EDate':'2008-12-31'})
            if timeseries.shape[0] > 0:
                if file_exist:
                    timeseries.to_csv("C:/Users/uqdkx/Desktop/Free_Float_Shares/"+test_ljc.iloc[i][2]+".csv", index = True, header = False, mode = 'a')
                else:
                    timeseries.to_csv("C:/Users/uqdkx/Desktop/Free_Float_Shares/"+test_ljc.iloc[i][2]+".csv", index = True)
                    file_exist = True
        except Exception as exc:
            if 'No data available for the requested date range' in str(exc):
                pass
            else:
                print(exc)

        time.sleep(1)
        try:
            timeseries, err = ek.get_data([test_ljc.iloc[i][2]], 
                                          ['TR.FreeFloat.Date', 'TR.FreeFloat'], 
                                          {'SDate':'2009-01-01', 'Frq':'D','EDate':'2019-05-07'})
            if timeseries.shape[0] > 0:
                if file_exist:
                    timeseries.to_csv("C:/Users/uqdkx/Desktop/Free_Float_Shares/"+test_ljc.iloc[i][2]+".csv", index = True, header = False, mode = 'a')
                else:
                    timeseries.to_csv("C:/Users/uqdkx/Desktop/Free_Float_Shares/"+test_ljc.iloc[i][2]+".csv", index = True)
                    file_exist = True
        except Exception as exc:
            if 'No data available for the requested date range' in str(exc):
                pass
            else:
                print(exc)
                
        time.sleep(1)
```
